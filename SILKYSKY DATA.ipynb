# Visualization

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)
```

# 1. Satisfaction by Gender - Bar Chart


```python
data.columns
```

```python
satisfaction_gender = data.groupby(['Gender', 'Satisfied']).size().unstack()
satisfaction_gender.plot(kind='bar', stacked=True)
plt.title("Satisfaction by Gender")
plt.xlabel("Gender")
plt.ylabel("count")
plt.legend(title= "Satisfied")
plt.show()
```

# 2. Satisfaction by Age Band - Bar Chart

```python
satisfaction_ageband =data.groupby(['Age Band', 'Satisfied']).size().unstack()
satisfaction_ageband.plot(kind='bar', stacked=True)
plt.title("Satisfaction by Age Band")
plt.xlabel("Age Band")
plt.ylabel("count")
plt.legend(title= "Satisfied")
plt.show()
```

# 3. Satisfaction vs Flight Distance - Scatter Plot

```python
colors = {'Y' : 'green', 'N' : 'red'}
plt.scatter(data['Flight Distance'],data['Age'], c=data['Satisfied'].map(colors), alpha=0.5)
plt.title("Satisfaction vs Flight Distance")
plt.xlabel("Flight Distance")
plt.ylabel("Age")
plt.grid(True)
plt.show()
```

# 4. Type of Travel vs Class vs Satisfaction - Stacked Bar Chart

```python
grouped = data.groupby(['Type of Travel', 'Class', 'Satisfied']).size().unstack().unstack()
grouped.plot(kind='bar', stacked=True)
plt.title("Satisfaction by Travel Type and Class")
plt.xlabel("Type of Travel / Class")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Satisfied")
plt.tight_layout()
plt.show()
```

# 5. Service Scores vs Satisfaction - Box Plots

```python
service_cols = ['Seat comfort', 'Inflight entertainment', 'On-board service',
                'Leg room service', 'Baggage handling', 'Checkin service',
                'Inflight service', 'Cleanliness']

for col in service_cols:
    sns.boxplot(data=data, x='Satisfied', y=col)
    plt.title(f"{col} by Satisfaction")
    plt.show()

# Optional: Violin Plot for visual smoothness
for col in service_cols:
    sns.violinplot(data=data, x='Satisfied', y=col, inner='quartile')
    plt.title(f"{col} by Satisfaction (Violin Plot)")
    plt.show()
```
